```python
import yt_dlp
import os

# Playlist URL
playlist_url = "https://music.youtube.com/playlist?list=PL6oZHyEtYrE7tCoJpvCu68kqYwzdJ0vZh&si=Ad9oN2bnwDi0076Q"

# Output folder on Google Drive
output_folder = "/content/drive/MyDrive/YouTube_Playlists"

# Create folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# yt-dlp options
ydl_opts = {
    'format': 'bestaudio/best',
    'ignoreerrors': True,
    'noplaylist': False,
    'outtmpl': f'{output_folder}/%(playlist_index)s - %(title)s.%(ext)s',
    'postprocessors': [{
        'key': 'FFmpegExtractAudio',
        'preferredcodec': 'mp3',
        'preferredquality': '192',
    }],
}

with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    ydl.download([playlist_url])
```
